```python
# derivatives first order
dy_dx = np.ones(2)
#print(dy_dx.shape)

def first_order(x_1,x_2):
  dy_dx[0] = 3*(x_1**2)*np.cos(x_2)
  dy_dx[1] = (x_1**3)*(-1*np.sin(x_2))

  return dy_dx.reshape(2,1)

```

```python
first_order(random_numbers_4[0], random_numbers_4[1] )
# 
#
```

```python

# convert numpy to torch 

x_torch = torch.from_numpy(random_numbers_4) # numpy to tensor 
x_torch.requires_grad_(True) # graident for backprop can be done with respect to inputs as True is on.
x_torch.retain_grad() # leaf tensor is a root of the tree 
print(x_torch)
#print(x_torch.shape)
```

```python
y_t = torch.zeros(1)
#print(y)
def Y_labels(x_1_t,x_2_t):
  y_t = (x_1_t**3)*(torch.cos(x_2_t))
  return y_t

```

```python
y_t = Y_labels(x_torch[0], x_torch[1]) # this maps with numpy version
```

```python
# Lets test for autograd works 

# first order derivative

y_t.backward() # gradients 
print(x_torch.grad)



```

```python
x_torch_1 = torch.from_numpy(random_numbers_4) # numpy to tensor 
x_torch_1.requires_grad_(True) # graident for backprop 
x_torch_1.retain_grad() # leaf tensor is a root of the tree 


torch.autograd.functional.hessian(Y_labels , (x_torch_1[0], x_torch_1[1] ))
```
