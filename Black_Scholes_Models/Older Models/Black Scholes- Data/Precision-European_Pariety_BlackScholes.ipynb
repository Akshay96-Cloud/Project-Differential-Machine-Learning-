```python
# Stock, Time, Rate ,Sigma -> 4 features (variables)
# Strike
# https://discuss.pytorch.org/t/tensor-stack-or-concatenate/34331/6 - column and concatinate 

X = torch.column_stack([Stock, Time, Rate, Sigma])
print(X)
print(X.shape) # m x n matrix 
```

```python
X.requires_grad_(True) # graident for backprop activated with requires_grad true
X.retain_grad() # leaf tensor is a root of the tree 

print(X.shape) 
```

```python
# non- divident paying asset. 
# P + S - C = K * e^-r(T-t)
# Set strike price = 100

Strike = 100 # fixed hyper paramters. 

# Our case this is a fixed paramter. 
```

Using Put-Call Pariety

```python
def euro_vanilla_call(x):
  d1 = (torch.log(x[0] / Strike) + (x[2] + 0.5 * x[3] ** 2) * x[1]) / (x[3] * torch.sqrt(x[1]))
  d2 = (torch.log(x[0] / Strike) + (x[2] - 0.5 * x[3] ** 2) * x[1]) / (x[3] * torch.sqrt(x[1])) # time x[1] with resepct to the time left not time. (- theta)
    
  call_t = (x[0] * torch.distributions.normal.Normal(0,1).cdf(d1) - Strike * torch.exp(-x[2] * x[1]) * torch.distributions.normal.Normal(0,1).cdf(d2))
  
  call = torch.nn.functional.relu(call_t)
  return call
```

```python
call_y = torch.zeros(size = (500 , 1) , dtype = torch.double) # samples

for i in range(X.shape[0]):
  call_y[i] = euro_vanilla_call(X[i]) # runs through each row computation. 
# numerical error (take absolute error of the prices and zero) -> take relu 
print(call_y) # M samples dimension all in all. 
```

```python
put_y = torch.zeros(size = (500 , 1))


put_y =  call_y -X[:,0].view(500,1) +Strike*torch.exp(-X[:,2].view(500,1)*X[:,1].view(500,1))
print(put_y)

# put from the put call pariety
```

```python
# P = K*e**(-rate*Time)*N(-d2) - S*N(-d1)

def euro_vanilla_put(x):
  d1 = (torch.log(x[0] / Strike) + (x[2] + 0.5 * x[3] ** 2) * x[1]) / (x[3] * torch.sqrt(x[1]))
  d2 = (torch.log(x[0] / Strike) + (x[2] - 0.5 * x[3] ** 2) * x[1]) / (x[3] * torch.sqrt(x[1]))
    
  put_t = (Strike*torch.exp(-x[2]*x[1])*torch.distributions.normal.Normal(0,1).cdf(-d2) - x[0]*torch.distributions.normal.Normal(0,1).cdf(-d1))

  put = torch.nn.functional.relu(put_t)
  return put
# https://pytorch.org/docs/stable/distributions.html - Normal( loc,scale)
#https://www.itl.nist.gov/div898/handbook/eda/section3/eda364.htm - loc and scale the same as mean and standard deviation.
# https://pytorch.org/docs/stable/generated/torch.log.html - returns natrual logarithem
# https://pytorch.org/docs/stable/generated/torch.sqrt.html - returns squart root
```

```python
put_original = torch.zeros(size = (500 , 1), dtype = torch.double) # samples

for i in range(X.shape[0]):
  put_original[i] = euro_vanilla_put(X[i]) # runs through each row computation. 

print(put_original) # M samples dimension all in all. 
```

```python
torch.isclose(put_y, put_original, atol = 1e-4) # 1e-5 is most values that are close.
#https://pytorch.org/docs/stable/generated/torch.isclose.html  # Close to absolute tolerance and relative torelnce check of data with true and false

# precision correct up to e-4 but a few erros after e-5 due to precision problems. 
# 4 numbers that are not close for 5
# But a few more for after 6. 
```
