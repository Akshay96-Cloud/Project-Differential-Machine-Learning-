```python
f_o = np.zeros(shape = (3,4))
f_o.shape
#print(f_o)
def first_order(x):
  f_o[:,0] = 3*(x[:,0]**2)*(x[:,1]**2)
  f_o[:,1] = 2*(x[:,0]**3)*x[:,1]
  f_o[:,2] = 9 * (x[:,2]**2)*(x[:,3]**2)
  f_o[:,3] = 6*(x[:,2]**3)*x[:,3]

  return f_o
```

```python

#print(s_o)

def second_order(x):
  s_o = np.zeros(shape = (3,4,4))

  # second orders relating to x1
  s_o[:,0,0] = 6*(x[:,0])*(x[:,1]**2)
  s_o[:,0,1] = 6*(x[:,0]**2)*(x[:,1])
  s_o[:,0,2] = 0 
  s_o[:,0,3] = 0 

  # Second Orders relating to x2. 
  s_o[:,1,0] = 6*(x[:,0]**2)*x[:,1]
  s_o[:,1,1] = 2*(x[:,0]**3)
  s_o[:,1,2] = 0 
  s_o[:,1,3] = 0 

  # Second Orders relating to x3 
  s_o[:,2,0] = 0 
  s_o[:,2,1] = 0 
  s_o[:,2,2] = 18*x[:,2]*(x[:,3]**2)
  s_o[:,2,3] = 18*(x[:,2]**2)*x[:,3]

  #Second Orders relating to x4
  s_o[:,3,0] = 0 
  s_o[:,3,1] = 0 
  s_o[:, 3, 2] = 18*(x[:,2]**2)*x[:,3]
  s_o[:, 3,3] =  6*(x[:,2]**3)

  return s_o

```

```python
# Factors for the second orders to normalise. 


# Do 4 factors for x1 variable. 

# 4 factors for x2 variable. 

# 4 factors for x3 varaible 

# 4 factors for x4 variable. 

# 4x4 16 factors but 6 are the same so 10 only. 
```

```python
# Diagonals

d2y_d2x1_factor = (x1_var)/ y_std # [0][0] position

d2y_d2x2_factor = (x2_var)/ y_std# [1][1] position

d2y_d2x3_factor = (x3_var)/ y_std # [2][2] position

d2y_d2x4_factor = (x4_var)/y_std # [3][3] position 

# Symmetries of 1 done

d2y_dx1dx2_factor = (x1_std*x2_std) / y_std # accounts for [0][1] and [1][0] position in the 4x4 as it is dx1dx2  and dx2dx1

d2y_dx1dx3_factor = (x1_std*x3_std) / y_std # accunts for [0][2] and the [2][0] position for dx1dx3 and dx3dx1

d2y_dx1dx4_factor = (x1_std*x4_std) / y_std # accounts for [0][3] and the [3][0] positin for dx1 and dx4 

# (2,3)  ( 2,4)  (4,3) remainng symmetires after 1. Done

d2y_dx2dx3_factor =  (x2_std* x3_std) / y_std # accounts for [2][1] and [1][2] positions for dx3dx2 and dx2dx3 

d2y_dx2dx4_factor =  (x2_std* x4_std) / y_std  # accounts for [3][1] and [1][3] position for dx2dx4 and dx4dx2

d2y_dx3dx4_factor =  (x3_std* x4_std) / y_std  # accounts for for [3][2] and [2][3] position for dx3dx4 and dx4dx3 





```

```python
# Global variables do not keep intalised variables outside functions intended use. Keep variables close inside the function. 

def second_order_normalise(p): 
  #Diagonal normalisation 
  p = np.zeros(shape = (x.shape[0],x.shape[1],x.shape[1])) # keep inside. 

  p[:,0,0] = d2y_d2x1_factor *s_o_un[:,0,0]
  p[:,1,1] = d2y_d2x2_factor * s_o_un[:,1,1]
  p[:,2,2] = d2y_d2x3_factor * s_o_un[:, 2,2]
  p[:,3,3] = d2y_d2x4_factor * s_o_un[:,3,3]


  # Normalise relation to x1 variable 
  p[:,0,1] = d2y_dx1dx2_factor* s_o_un[:,0,1]
  p[:,1,0] = d2y_dx1dx2_factor* s_o_un[:,1,0] # Both representing dx1dx2 and dx2dx1 normalisisation 

  p[:,0,2] = d2y_dx1dx3_factor * s_o_un[:,0,2]
  p[:,2,0] = d2y_dx1dx3_factor * s_o_un[:,2,0] # both represent dx1dx3 and dx3dx1 

  p[:,0,3] = d2y_dx1dx4_factor*s_o_un[:,0,3] 
  p[:,3,0] = d2y_dx1dx4_factor*s_o_un[:,3,0]

  # Normalise last combinations 
  p[:, 2, 1] = d2y_dx2dx3_factor* s_o_un[:, 2, 1]
  p[:, 1, 2] = d2y_dx2dx3_factor *s_o_un[:, 1,2] # dx2dx3 and dx3dx2

  p[:, 3, 1 ] = d2y_dx2dx4_factor *s_o_un[:,3,1]
  p[:, 1, 3 ] = d2y_dx2dx4_factor* s_o_un[:,1, 3] #dx4dx2 and dx2dx4 


  p[:, 3, 2] = d2y_dx3dx4_factor*s_o_un[:,3 ,2]
  p[:, 2 ,3] - d2y_dx3dx4_factor*s_o_un[:,2,3]

  return p 

```

```python
s_o_norm = second_order_normalise(s_o_un)
print(s_o_norm)
```
